```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
%matplotlib widget
x = np.arange(1,10)
z = np.sin(x)
figure = TheArtist(latex=False,n_cols = 2,n_rows = 2)
figure.plot(x, z, 0, 0, color='darkblue', linewidth=1, linestyle='-', marker='o')
figure.scatter(x, z, 1, 0, marker='s')
figure.plot(x, z, 1, 1, color='darkblue', linewidth=1, linestyle='-', marker='o')
figure.scatter(x, z, 0, 1, marker='s')
figure.set_labels(['Eje 1', 'Eje V'], 0, 0)
```

```python
iris = load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
df.head()
```

```python
figure = TheArtist(latex=False,n_cols=1, n_rows=1,)
figure.plot(df.index,df["sepal length (cm)"], 0, 0, marker='s')
```
