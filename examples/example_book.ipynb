```python
from TheArtist import TheArtist
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
%matplotlib widget
x = np.arange(1,10)
z = np.sin(x)
figure = TheArtist(latex=False)
figure.generate_figure_environment(cols=2, rows=2, fig_width_pt=384, ratio='golden', regular=True)
figure.plot_lines(x, z, 0, 0, color='darkblue', linewidth=1, linestyle='-', marker='o')
figure.plot_scatter(x, z, 1, 0, marker='s')
figure.plot_lines(x, z, 1, 1, color='darkblue', linewidth=1, linestyle='-', marker='o')
figure.plot_scatter(x, z, 0, 1, marker='s')
figure.set_labels(['Eje 1', 'Eje V'], 0, 0)
figure.set_labels(['Eje 2', None], 0, 0)
```

```python
iris = load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
df.head()
```

```python
figure = TheArtist(latex=False)
figure.generate_figure_environment(cols=1, rows=1, fig_width_pt=384, ratio='golden', regular=True)
figure.plot_scatter(df.index,df["sepal length (cm)"], 0, 0, marker='s')
```
